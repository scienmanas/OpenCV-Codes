```python
# Splitting and merging the colour channels

img_NZ_bgr = cv2.imread("New_Zealand_Lake.jpg", cv2.IMREAD_COLOR)


b, g, r = cv2.split(img_NZ_bgr)

# Show Channels

# figsize = [width, height]
plt.figure(figsize=[20,5])

plt.subplot(141)
plt.title('Red Channel')
plt.imshow(r, cmap='gray')

plt.subplot(142)
plt.title('Green Channel')
plt.imshow(g, cmap='gray')

plt.subplot(143)
plt.title('Blue Channel')
plt.imshow(b, cmap='gray')


imgMerged = cv2.merge((b,g,r))

plt.subplot(144)
plt.title('All Channels')
plt.imshow(imgMerged[:,:,::-1])


```

```python
'''
Converting into different colout spaces.

Note that standard image is a 24 bit  per pixel RGB image, 8 - 8 bit colour space the first three bytes being rgb and the the next 3 bytes another pixel
'''


# Converting into different color spaces

# Changing from BGR to RGB

img_NZ_rgb = cv2.cvtColor(img_NZ_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_NZ_rgb)
```

```python
# Chnaging to HSV color space

img_hsv = cv2.cvtColor(img_NZ_bgr, cv2.COLOR_BGR2HSV)

h,s,v = cv2.split(img_hsv)

plt.figure(figsize=[20,5])

plt.subplot(141)
plt.imshow(h, cmap="gray")
plt.title("Hue channel")

plt.subplot(142)
plt.imshow(s,  cmap="gray")
plt.title("Saturation channel")

plt.subplot(143)
plt.imshow(v, cmap='gray')
plt.title('Value Channel')

plt.subplot(144)
plt.imshow(img_NZ_rgb)
plt.title("Original")
```

```python
# Modyfing individual channels

h_new = h + 10
img_NZ_merged = cv2.merge([h_new, s, v])
img_NZ_rgb = cv2.cvtColor(img_NZ_merged, cv2.COLOR_HSV2RGB)

# Plotting

plt.figure(figsize=[20, 5])
plt.subplot(141)
plt.imshow(h,  cmap='gray')
plt.title('Original Hue channel')

plt.subplot(142)
plt.imshow(s, cmap='gray')
plt.title("Saturation")

plt.subplot(143)
plt.imshow(v, cmap="gray")
plt.title("Value")

plt.subplot(144)
plt.imshow(img_NZ_rgb)
plt.title("Modified")
```

```python
# Saving images

cv2.imwrite("New_Zealand_Lake_Saved.png", img_NZ_bgr)

Image(filename='New_Zealand_Lake_Saved.png')
```
