```python
# Basic Image manipulations

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import Image
%matplotlib inline
```

```python
# Modifying the pixels

cb_img = cv2.imread('checkerboard_18x18.png',0)

print(cb_img)
plt.imshow(cb_img, cmap='gray')
# cv2.imshow('Image', cb_img)
```

```python
# Accessing individual pixels

print(cb_img[0,0])

print(cb_img[0,5])
print(cb_img[0,6])
```

```python
# Modifying image pixels

cb_img_copy = cb_img.copy()
cb_img_copy[2,2] = 200
cb_img_copy[2,3] = 200
cb_img_copy[3,2] = 200
# cb_img_copy[3,3] = 200

plt.imshow(cb_img_copy, cmap='gray')
print(cb_img_copy)
```

```python
# Cropping images
'''
Cropping is simply achieved by slecting a specific (pixel) region of the image
'''

img_NZ_bgr = cv2.imread("New_Zealand_Boat.jpg", cv2.IMREAD_COLOR)
img_NZ_bgr = cv2.cvtColor(img_NZ_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_NZ_bgr)


```

```python
# Now Cropping the image

cropped_region = img_NZ_bgr[200:400, 300:600]
plt.imshow(cropped_region)
```
